```python
import numpy as np
import matplotlib.pyplot as plt
from my_tools import *

np.random.seed(0)
```

```python
a = np.array([1/2, 1/4, -1/4, 1/4, 3/4], dtype=np.float32)
b = np.array([2, 4, 3, 3, 0], dtype=np.float32)

def J(x):
#     функция, которую минимизируем
    return np.dot(a,x)**4 + 2 * np.dot(x-b, x-b) 

def der_J(x):
    # реальная производная с 4, но разницы нет
    return ((np.dot(a,x)**3)*a + x-b) #*4
```

```python
plt.rcParams['text.usetex'] = True

num_lines = 50

start_points = [np.random.rand(5)*10 + (-a+b) for i in range(num_lines)]
```

```python
eps = 0.01 # точность
# max_iter макс кол-во итераций
# alpha начальный шаг в Армихо или просто константный
c = 1/4 # Коэффициент в армихо
```

```python
# 1

max_iter = 30
alpha = 1

distance_curves(start_points, J, der_J, 
                alpha, eps, max_iter, 
                conj_method="DY", step_method="armijo", c=c, fig_size=[7, 5],
                a=a, b=b, cond="conj")

# plt.savefig('../prac_tex/imgs/img1.png')
```
